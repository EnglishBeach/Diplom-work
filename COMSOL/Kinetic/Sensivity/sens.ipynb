```python
compounds = ['D', 'DH', 'Q', 'QHH', 'QHD', 'prod']
frames0=[]
frames1=[]
for compound in compounds:
    df = pd.read_csv(f'{compound}.csv', delimiter=',', header=4)
    a = {
        'Ke': '1000000000',
        'KH': '1000000000',
        'Kr': '1000000000',
        'Kp': '0.001',
        'KqH': '2000',  # 'Kdisp': '1000000000',
        'Kph': '0.00001',
        'Ks': '2',
        'Kd': '0.05',  # 'Kc': '1',
        'KrD': '1000000000'
    }
    CONSTANTS = {key: np.float64(value) for key, value in a.items()}
    replaced_columns = {
        col: col.replace('fsens_', '')
        for col in df.columns
        if 'Time' not in col
    }
    replaced_columns.update({
        '% Time (s)': 'time',
    })
    df = df.rename(columns=replaced_columns)
    for col in CONSTANTS:
        df[col] = df[col] * CONSTANTS[col]
    df['compound'] = compound
    
    slice0 = df.iloc[len(df) // 4]
    frames0.append(pd.DataFrame(slice0).T)
    
    slice1 = df.iloc[-1]
    frames1.append(pd.DataFrame(slice1).T)
    
result0 = pd.concat(frames0)
result0.drop(columns=['time'],inplace=True)

result1 = pd.concat(frames1)
result1.drop(columns=['time'],inplace=True)
```

```python
melt_result0 = result0.melt('compound',var_name ='consts' ,value_name='values',)
melt_result1 = result1.melt('compound',var_name ='consts' ,value_name='values',)
```

```python
fig = plyx.line(
    data_frame=melt_result0,
    x="compound",
    y="values",
    color='consts',
    markers=True,
    # hovertemplate="Аргумент: %{x}<br>Функция: %{y}",
    title="layout.hovermode='x'"
)
fig.update_traces(
     hovertemplate=None
)
fig.update_layout(
    margin=dict(l=0, r=0, b=0, t=0, pad=0),
    legend=dict(x=-0.1, y=1, xanchor="left"),
    hovermode="x",
    
)
```
