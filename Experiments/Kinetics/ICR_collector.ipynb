```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ocm = pd.read_excel(r'графики.xlsx', sheet_name=0)
ocm = pd.DataFrame(
    {'t': [0] + ocm.iloc[:, 0].to_list(), 'p': [0] + ocm.iloc[:, 1:3].mean(axis=1).to_list()}
)
peta = pd.read_excel(r'графики.xlsx', sheet_name=1)
peta = pd.DataFrame(
    {'t': [0] + peta.iloc[:, 0].to_list(), 'p': [0] + peta.iloc[:, 1:3].mean(axis=1).to_list()}
)
dmeg = pd.read_excel(r'графики.xlsx', sheet_name=2)
dmeg = pd.DataFrame(
    {'t': [0] + dmeg.iloc[:, 0].to_list(), 'p': [0] + dmeg.iloc[:, 1:3].mean(axis=1).to_list()}
)
```

```python
ocm.to_excel(r'results\ocm.xlsx')
peta.to_excel(r'results\peta.xlsx')
dmeg.to_excel(r'results\dmeg.xlsx')
```

```python
def get_m(P,M0=2.8):
    return -M0 *(-1 + P)
```

```python
plt.plot(ocm['t'],ocm['p'],label='ocm')
plt.plot(peta['t'],peta['p'],label='peta')
plt.plot(dmeg['t'],dmeg['p'],label='dmeg')
plt.legend()
```
